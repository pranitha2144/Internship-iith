```
import numpy as np
import os
import time
from keras.applications import ResNet50
from keras.preprocessing import image
from keras.layers import GlobalAveragePooling2D,Dense, Dropout,Activation,Flatten

from keras.applications.imagenet_utils import preprocess_input
from keras.layers import Input
from keras.models import Model
from keras.utils import np_utils
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
```

```
img_path='/content/drive/My Drive/data/grade0/grade0-001.jpg'
img=image.load_img(img_path,target_size=(224,224))
x=image.img_to_array(img)
print(x.shape)
x=np.expand_dims(x,axis=0)
print(x.shape)
x=preprocess_input(x)
print(x.shape)
```

```
data_path='/content/drive/My Drive/data_ouput/train'
data_dir_list=os.listdir(data_path)
img_data_list=[]
for dataset in data_dir_list:
  img_list=os.listdir(data_path+'/'+dataset)
  print('loaded images of dataset '+'{}\n'.format(dataset))
  for img in img_list:
    img_path=data_path+'/'+dataset+'/'+img
    img=image.load_img(img_path,target_size=(224,224))
    x=image.img_to_array(img)
    x=np.expand_dims(x,axis=0)
    x=preprocess_input(x)
    print(x.shape)
    img_data_list.append(x)
```

```
img_data=np.array(img_data_list)
print(img_data.shape)
img_data=np.rollaxis(img_data,1,0)
print(img_data.shape)
img_data=img_data[0]
print(img_data.shape)
```

```
num_classes=4
no_of_samples=img_data.shape[0]
labels = np.ones((no_of_samples,),dtype='int64')
```

```
labels[0:200]=0
labels[200:400]=1
labels[400:600]=2
labels[600:]=3
```

```
names=['grade1','grade0','grade2','grade3']
```

```
Y=np_utils.to_categorical(labels,num_classes)
Y
Y.shape
```

```
data_path='/content/drive/My Drive/data_ouput/val'
data_dir_list=os.listdir(data_path)
img_data_list=[]
for dataset in data_dir_list:
  img_list=os.listdir(data_path+'/'+dataset)
  print('loaded images of dataset '+'{}\n'.format(dataset))
  for img in img_list:
    img_path=data_path+'/'+dataset+'/'+img
    img=image.load_img(img_path,target_size=(224,224))
    x=image.img_to_array(img)
    x=np.expand_dims(x,axis=0)
    x=preprocess_input(x)
    print(x.shape)
    img_data_list.append(x)
```

```
img_data_val=np.array(img_data_list)
print(img_data_val.shape)
img_data_val=np.rollaxis(img_data_val,1,0)
print(img_data_val.shape)
img_data_val=img_data_val[0]
print(img_data_val.shape)
```

```
num_classes=4
no_of_samples=img_data_val.shape[0]
labels = np.ones((no_of_samples,),dtype='int64')
```

```
labels[0:25]=0
labels[25:50]=1
labels[50:75]=2
labels[75:]=3
```

```
Y_val=np_utils.to_categorical(labels,num_classes)
Y_val.shape
```

```
x_train,y_train=shuffle(img_data,Y,random_state=20)
```

```
x_val,y_val=shuffle(img_data_val,Y_val,random_state=20)
```

```
X_train,X_test,Y_train,Y_test=train_test_split(img_data,Y,test_size=0.2,random_state=2)
```

```
print(X_train.shape)
print(Y_train.shape)
print(Y_test.shape)
print(X_test.shape)
```

```
image_input=Input(shape=[224,224,3])
model=ResNet50(input_tensor=image_input,include_top=True,weights='imagenet')
```

```
model.summary()
```

```
from keras import regularizers
last_layer=model.get_layer('avg_pool').output
fc_1=Dense(32,activation='relu',kernel_regularizer=regularizers.l2(0.0001))(last_layer)
dropout=Dropout(0.4)(fc_1)
out=Dense(num_classes,activation='softmax',name='output_layer')(dropout)
custom_resnet_model=Model(inputs=image_input,outputs=out)
custom_resnet_model.summary()
```

```
for layer in custom_resnet_model.layers[:-3]:
  layer.trainable=False
#custom_resnet_model.layers[-2].trainable=True
custom_resnet_model.layers[-1].trainable
```

```
custom_resnet_model.summary()
```

```
custom_resnet_model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```
t=time.time()
hist=custom_resnet_model.fit(x_train,y_train,epochs=12,shuffle=True, validation_data=(x_val,y_val))
print("training time is %s"%(t-time.time()))
```
